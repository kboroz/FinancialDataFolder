```python
import pandas as pd

# Import the dataframe

#url2 = 'https://raw.githubusercontent.com/kboroz/MSFE-DataFolder/main/Returns_Stocks%2BBonds.csv'

#url2 = 'https://raw.githubusercontent.com/kboroz/MSFE-DataFolder/main/Absolute-Values_Stocks_Bond.csv'

url2 = 'https://raw.githubusercontent.com/kboroz/MSFE-DataFolder/main/Stocks_Bonds_Final.csv'

stocks_bonds = pd.read_csv(url2)

# Ensure the Date column is in datetime format and change its format to "year-month-day"
stocks_bonds['Date'] = pd.to_datetime(stocks_bonds['Date'], format='%d.%m.%Y').dt.strftime('%Y-%m-%d')

# Display the DataFrame to verify the date format change
print(stocks_bonds)
```

```python
import pandas as pd

# Read the CSV files into DataFrames
df1 = stocks_bonds
df2 = crypto

# Convert "Date" column to datetime format
df1['Date'] = pd.to_datetime(df1['Date'])
df2['Date'] = pd.to_datetime(df2['Date'])

# Filter df2 to include only dates present in df1
df2_filtered = df2[df2['Date'].isin(df1['Date'])]

# Merge the DataFrames on the "Date" column
merged_df = pd.merge(df1, df2_filtered, on='Date')

# Export merged DataFrame to CSV
merged_df.to_csv('merged_file.csv', index=False)

# Display the merged DataFrame
print(merged_df)

```

```python
# Sort the DataFrame by the "Date" column in descending order
merged_df_sorted_desc = merged_df.sort_values(by='Date', ascending=False)

# Then extract every fifth entry from the sorted DataFrame
every_fifth_entry_sorted_desc = merged_df_sorted_desc.iloc[::5]

# Display the extracted entries
print(every_fifth_entry_sorted_desc)
```

```python
every_fifth_entry_sorted_desc
```

```python
# Load data into DataFrame
every_fifth_entry_sorted_desc = pd.DataFrame(every_fifth_entry_sorted_desc)

# Convert numbers to US-format (comma as decimal separator) to float
for col in every_fifth_entry_sorted_desc.columns[1:]:
    every_fifth_entry_sorted_desc[col] = every_fifth_entry_sorted_desc[col].str.replace(',', '.').astype(float)

# Convert DataFrame to CSV with US-format
every_fifth_entry_sorted_desc_output = every_fifth_entry_sorted_desc.to_csv(index=False)

print(every_fifth_entry_sorted_desc_output)
```

```python
# Export merged DataFrame to CSV
every_fifth_entry_sorted_desc.to_csv('every_fifth_entry_sorted_desc_output.csv', index=False)
```

```python
import numpy as np

df = pd.DataFrame(df)

# Set the Date column as the index (if it's not already)
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)

# Calculate the log-returns
log_returns = np.log(df.shift(1) / df)

print(log_returns)
```

```python
log_returns
```

```python
# Calculate the correlation matrix for the log-returns
correlation_matrix = log_returns.corr()

print(correlation_matrix)
```

```python
# Calculate the covariance matrix for the log-returns
covariance_matrix = log_returns.cov()

print(covariance_matrix)
```
